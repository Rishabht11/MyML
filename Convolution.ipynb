```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import skimage.io as io
```

```python
img = cv2.imread('/home/abc/Desktop/download.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print img.shape, img.dtype
plt.figure(0,figsize=(7,7))
plt.imshow(img[:, :], cmap='gray')
```

```python
kernel = np.array([[1.,2.,1.],
                 [0.,0.,0.],
                 [-1.,-2.,-1.]])
print kernel
```

```python
new_image = np.zeros((img.shape[0]-kernel.shape[0]+1, img.shape[1]-kernel.shape[1]+1))
```

```python
for ix in range(new_image.shape[0]):
    for iy in range(new_image.shape[1]):
        im_patch = img[ix:ix+kernel.shape[0],iy:iy+kernel.shape[1]]
        h_prod = im_patch * kernel
        new_image[ix, iy] = max(0,h_prod.sum()) #ReLU(passing the prod, if it is greater tha 0)
plt.imshow(new_image.astype(np.uint8),cmap='gray')
```
